```python
# Regular EDA and plotting libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 

# Player Name Constant, Change if necessary
PLAYER_NAME = "LeBron James"


# Read data
games = pd.read_csv('data/games.csv')
games_details = pd.read_csv('data/games_details.csv')
ranking = pd.read_csv('data/ranking.csv')
```

```python
games_details = games_details.drop_duplicates(subset=["GAME_ID", "PLAYER_NAME"])
games_details.head()
```

```python
bron = games_details.loc[games_details["PLAYER_NAME"] == PLAYER_NAME]
bron = bron.drop(["TEAM_ID", "TEAM_CITY", "PLAYER_ID", "PLAYER_NAME", "COMMENT"], axis=1)
bron.head()
```

```python
# Remove Season 2020 as to exclude COVID-19 season
games_before_season_2020 = games.loc[games['SEASON'] < 2020]
```

```python
games_date = games_before_season_2020[["GAME_DATE_EST", "GAME_ID", "SEASON"]]
stats = bron.merge(games_date, on="GAME_ID", how="left")
seasonal_stats = stats.groupby("SEASON").sum()/stats.groupby("SEASON").count()
seasonal_stats.head()
```

# LeBron Stats Visualization

# LeBron's Team Winning Percentage

```python
bron_details = pd.merge(games, games_details[games_details.PLAYER_NAME == PLAYER_NAME], on = "GAME_ID")
bron_details["home"] = bron_details["TEAM_ID"] == bron_details["TEAM_ID_home"]
bron_details["WIN"] = (bron_details["home"] & (bron_details["PTS_home"] > bron_details["PTS_away"])) | ((bron_details["home"] == False) & (bron_details["PTS_home"] < bron_details["PTS_away"]))
```

```python
fig, ax = plt.subplots(figsize=(10,10))
fig.suptitle("LeBron Team Percentage", fontsize=20, color = "black")

ax.pie(x = bron_details.groupby(["TEAM_ABBREVIATION"]).count().home, labels = ["CLE", "LAL", "MIA"], autopct="%.1f%%", pctdistance=0.5, wedgeprops=dict(edgecolor='w'))
```

```python
plt.figure(figsize=(10,10))

plt.pie(x=bron_details.groupby("WIN").count().home, labels = ["LOSE", "WIN"], autopct="%.1f%%", pctdistance=0.5, wedgeprops=dict(edgecolor='w'))
plt.title("LeBron's Team Win Percentage", fontsize=20, color = "black")
```
